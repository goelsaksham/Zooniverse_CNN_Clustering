# Importing Libraries

Importing the various libraries for loading images and applying data preprocessing

```python
import numpy as np
import pandas as pd
import tensorflow as tf
import os
import glob
from skimage.transform import resize
import cv2
import PIL
import matplotlib.pyplot as plt
%matplotlib inline
```

## Data Preprocessing

* Resizing all image channels to 224 x 224
* Stacking channels to form Images of shape (224 x 224 x 3)
* Mean Subtracting Images and normalizing to a range of (-1, 1)

```python
# Load the image
image_obj = PIL.Image.open('../../data/temp_data/spiral_test.jpg')
image = np.array(image_obj)
print(image.shape)
```

```python
# Display the image
plt.imshow(image)
```

```python
# Resizing the image
resized_image = resize(image, (224, 224, 3))
# Displaying the image
plt.imshow(resized_image)
```

```python
resized_image = resized_image - 0.5
resized_image = resized_image/0.5
plt.imshow(resized_image)
```
