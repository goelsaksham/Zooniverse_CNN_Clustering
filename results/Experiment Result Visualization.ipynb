# Experiment Results Visualization

This notebook helps in plotting the results of various experiments conducted for the part of the galaxy clustering research

### Importing the required libraries and modules

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set()
```

### Loading the data

```python
preprocessing_results = pd.read_csv(f'preprocessing_results.csv')
kmeans_initialization_results = pd.read_csv(f'kmeans_initialization_results.csv')
agglomerative_clustering_results = pd.read_csv(f'agglomerative_clustering_results.csv')
```

```python
preprocessing_results
```

```python
kmeans_initialization_results
```

```python
agglomerative_clustering_results
```

### Data Preprocessing

```python
cm = sns.light_palette("green", as_cmap=True)
preprocessing_results.style.background_gradient(cmap=cm)
```

```python
df = pd.pivot_table(preprocessing_results, 'Clustering Accuracy', 'Feature Preprocessing', 'Image Preprocessing')
fig, ax = plt.subplots(1, 1, figsize=(10, 7))
sns.heatmap(df, cmap=sns.cubehelix_palette(15), annot=True, fmt='.8g', linewidths=3, ax=ax, annot_kws={"size":20})
ax.set_title('Data Preprocessing Clustering Acccuracy Comparison')
```

```python
df = pd.pivot_table(preprocessing_results, 'Clustering NMI', 'Feature Preprocessing', 'Image Preprocessing')
fig, ax = plt.subplots(1, 1, figsize=(10, 7))
sns.heatmap(df, cmap=sns.cubehelix_palette(15), annot=True, fmt='.8g', linewidths=3, ax=ax, annot_kws={"size":20})
ax.set_title('Data Preprocessing Clustering NMI Comparison')
```

```python
df = pd.pivot_table(preprocessing_results, 'Clustering Loss', 'Feature Preprocessing', 'Image Preprocessing')
fig, ax = plt.subplots(1, 1, figsize=(10, 7))
sns.heatmap(df, cmap=sns.cubehelix_palette(15), annot=True, fmt='.8g', linewidths=3, ax=ax, annot_kws={"size":20})
ax.set_title('Data Preprocessing Clustering Loss Comparison')
```

```python
df = pd.pivot_table(preprocessing_results, 'Reconstruction Loss', 'Feature Preprocessing', 'Image Preprocessing')
fig, ax = plt.subplots(1, 1, figsize=(10, 7))
sns.heatmap(df, cmap=sns.cubehelix_palette(8), annot=True, fmt='.8g', linewidths=3, ax=ax, annot_kws={"size":20})
ax.set_title('Data Preprocessing Reconstruction Loss Comparison')
```

### Semi-Supervised Clustering

```python
cm = sns.light_palette("green", as_cmap=True)
kmeans_initialization_results.style.background_gradient(cmap=cm)
```

```python
fig, ax = plt.subplots(1, 1)
sns.scatterplot(x='Clustering Accuracy', y='Clustering NMI', hue='K-Means Initialization', 
                data=kmeans_initialization_results, s=200, alpha=0.8, marker='X')
ax.set_title('Comparison of K-Means Center Initialization Technique')
```

```python
fig, ax = plt.subplots(1, 1)
sns.scatterplot(x='Clustering Loss', y='Reconstruction Loss', hue='K-Means Initialization', 
                data=kmeans_initialization_results, s=200, alpha=0.8, marker='X')
ax.set_title('Comparison of K-Means Center Initialization Technique')
```

### Agglomerative Clustering

```python
cm = sns.light_palette("green", as_cmap=True)
agglomerative_clustering_results.style.background_gradient(cmap=cm)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(7, 5))
sns.scatterplot(x='Clustering Accuracy', y='Clustering NMI', hue='Agglomerative Clustering', 
                data=agglomerative_clustering_results, s=200, alpha=0.8, marker='X')
ax.set_title('Comparison of Agglomerative Clustering')
```

```python
fig, ax = plt.subplots(1, 1, figsize=(7, 5))
sns.scatterplot(x='Clustering Loss', y='Reconstruction Loss', hue='Agglomerative Clustering', 
                data=agglomerative_clustering_results, s=200, alpha=0.8, marker='X')
ax.set_title('Comparison of Agglomerative Clustering')
```

## All Result Comparison

```python
df = pd.concat([preprocessing_results, kmeans_initialization_results, agglomerative_clustering_results])
df
```

```python
df['Model'] = df['Agglomerative Clustering']
```

```python
df['Model'][~pd.isna(df['Feature Preprocessing'])] = df[['Feature Preprocessing', 'Image Preprocessing']].apply(lambda x: f"{x['Feature Preprocessing']}_{x['Image Preprocessing']}", axis=1)
```

```python
df['Model'][~pd.isna(df['K-Means Initialization'])] = df['K-Means Initialization'][~pd.isna(df['K-Means Initialization'])]
```

```python
df
```

```python
df.drop(labels=['Agglomerative Clustering', 'K-Means Initialization', 'Feature Preprocessing', 'Image Preprocessing'], 
        axis=1, inplace=True)
```

```python
df
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 5))
sns.scatterplot(x='Clustering Accuracy', y='Clustering NMI', hue='Model', 
                data=df, s=200, alpha=0.8, marker='X')
ax.set_title('Comparison of All Experiments')
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 5))
sns.scatterplot(x='Clustering Loss', y='Reconstruction Loss', hue='Model', 
                data=df, s=200, alpha=0.8, marker='X')
ax.set_title('Comparison of All Experiments')
```
